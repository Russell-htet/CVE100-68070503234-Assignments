```python
class Car:
    def __init__(self,model, rent_per_day, available=True):
        self.model = model
        self.rent_per_day = rent_per_day
        self. available = True
        
    def rent(self, days):
        if self.available:
            self.available = False
            total_cost = self.rent_per_day * days
            print("Car rented. Total cost = " + str(total_cost))
        else:
            print("Car not available")
        
    def return_car(self):
        self.available = True
        print("Car returned")

    def display(self):
        print("Model:" + str(self.model) + ",Rent per day: " + str(self.rent_per_day) + ",Status:" + str(self.available))
    
car = Car("BMW M4",100)
car.rent(5)
car.display()
car.return_car()
car.display()


```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def display_info(self):
        print("Name:", self.name, "Age:", self.age)

class Student(Person):
    def __init__(self, name, age):
        super().__init__(name, age)
        self.marks = {}

    def add_marks(self, subject, score):
            if subject not in self.marks:
                self.marks[subject] = score
                print(f"Added {subject} = {score}")
            else:
                self.marks[subject] = score
                print(f"Updated {subject} = {score}")

    def display_marks(self):
        print("Student Marks:")
        for subject, score in self.marks.items():
            print(f"{subject} : {score}")


    def average(self):
        if not self.marks:
            return 0
        total_score = sum(self.marks.values())
        number_of_subjects = len(self.marks)
        return total_score / number_of_subjects
    
student1 =Student("Russell", 18)
student1.display_info()

student1.add_marks("Maths", 89)
student1.add_marks("Physics", 99)

student1.display_marks()

average_score = student1.average()
print(f"Average Score: {average_score}")
```
